Stacking Can add up the array vertically or horizontal

```python
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
np.hstack((a4,a5,a4,a5))#Horizontal Stack
np.vstack((a4,a5,a4,a5))#Verical Stack
```

Splitting Horizontally and vertically split

```python
np.hsplit(a5,4)#horizontal split
np.vsplit(a5,3)#Vertical Split
```
