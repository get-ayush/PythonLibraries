```python
# Write a function that can return the track record of 2 teams against each other
def get_track_record(team1, team2):
    # Filter matches where both teams played against each other
    mask = ((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))
    temp_df = ipl[mask]

    total_matches = temp_df.shape[0]
    if total_matches == 0:
        return f'No matches found between {team1} and {team2}.'

    team1_wins = temp_df[temp_df['WinningTeam'] == team1].shape[0]
    team2_wins = temp_df[temp_df['WinningTeam'] == team2].shape[0]
    draws_or_no_result = total_matches - (team1_wins + team2_wins)

    result = {
        'Total Matches': total_matches,
        team1: team1_wins,
        team2: team2_wins,
        'No Result/Draw': draws_or_no_result
    }
    return result

# Example usage:
print(get_track_record('Chennai Super Kings', 'Mumbai Indians'))
```
